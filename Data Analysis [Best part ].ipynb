# Mean rating by Player

```python
#Full_Data=Full_Data.reset_index()

mean_player_rating=Full_Data[(Full_Data.team_name == 'Arsenal')].groupby(['player_name'])['rating'].mean().dropna()
mean_player_rating
```

```python
mean_player_rating_plot = mean_player_rating.plot(kind='bar', title="Mean Player Rating",figsize =(12,8))
plt.xticks(rotation=90)
plt.xlabel('Player_Name')
plt.ylabel('Rate')
mean_player_rating_plot
```
